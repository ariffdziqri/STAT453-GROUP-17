```python
import json
import pandas as pd
import glob

# Load all JSONL chunk files
files = glob.glob("data/chunks/*.jsonl")
records = []
for f in files:
    with open(f) as fh:
        for line in fh:
            records.append(json.loads(line))

df = pd.DataFrame(records)
df.head()
```

```python
print(df.shape)          # (rows, cols)
print(df.dtypes)
df.describe()            # stats on token_count, chunk_index, etc.
```

```python
# chunks per ticker
df.groupby("ticker").size()

# chunks per section
df.groupby("section_label").size().sort_values(ascending=False)

# pivot: tickers × sections
df.pivot_table(index="ticker", columns="section_label", values="chunk_index", aggfunc="count", fill_value=0)
```

```python
import matplotlib.pyplot as plt

df["token_count"].hist(bins=40)
plt.title("Token count per chunk")
plt.xlabel("tokens")
plt.show()

# by section
df.boxplot(column="token_count", by="section_label", rot=45, figsize=(10,5))
plt.tight_layout()
plt.show()
```
