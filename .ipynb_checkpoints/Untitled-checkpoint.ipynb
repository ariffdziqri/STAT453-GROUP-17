```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent": "Plokkage aismadi@wisc.edu"}

# Load the ticker -> CIK mapping from SEC
url = "https://www.sec.gov/files/company_tickers.json"
response = requests.get(url, headers=headers)
data = response.json()

# Find Tesla
for entry in data.values():
    if entry["ticker"].upper() == "TSLA":
        cik = str(entry["cik_str"]).zfill(10)
        print("CIK:", cik)
        break
```

```python
filings_url = f"https://data.sec.gov/submissions/CIK{cik}.json"
filings = requests.get(filings_url, headers=headers).json()

# Peek at company info
print("Company:", filings["name"])
print("SIC:", filings["sic"])
print("Total filings:", len(filings["filings"]["recent"]["form"]))
```

```python
forms = filings["filings"]["recent"]["form"]
accessions = filings["filings"]["recent"]["accessionNumber"]
dates = filings["filings"]["recent"]["filingDate"]

for i, form in enumerate(forms):
    if form == "10-K":
        accession = accessions[i].replace("-", "")
        date = dates[i]
        print(f"Found 10-K filed on: {date}")
        print(f"Accession number: {accession}")
        break
```

```python
cik_int = int(cik)  # remove leading zeros for the URL
index_url = f"https://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK={cik_int}&type=10-K&dateb=&owner=include&count=1&search_text="

index_response = requests.get(index_url, headers=headers)
print("Status code:", index_response.status_code)
print("First 500 chars:")
print(index_response.text[:500])
```

```python
doc_url = f"https://www.sec.gov/Archives/edgar/data/{cik_int}/{accession}/{accession}-index.htm"
doc_response = requests.get(doc_url, headers=headers)

print("Status code:", doc_response.status_code)
print("\nPreview:")
print(doc_response.text[:1000])
```

```python
headers = {"User-Agent": "Plokkage myemail@wisc.edu"}
cik = "0001318605"

# Get filing list
filings = requests.get(f"https://data.sec.gov/submissions/CIK{cik}.json", headers=headers).json()

# Get latest 10-K accession number
forms = filings["filings"]["recent"]["form"]
accessions = filings["filings"]["recent"]["accessionNumber"]

for i, form in enumerate(forms):
    if form == "10-K":
        accession = accessions[i].replace("-", "")
        print("Accession:", accession)
        break
```

```python
# The index page lists every file in this filing
index_url = f"https://www.sec.gov/Archives/edgar/data/{int(cik)}/{accession}/{accession}-index.json"
index_data = requests.get(index_url, headers=headers).json()

# Find the main 10-K HTML document (largest .htm file)
for doc in index_data["directory"]["item"]:
    if doc["type"] == "10-K":
        main_doc = doc["name"]
        print("Main document:", main_doc)
        break
```

```python
# Step 1: print the raw accession number to see exactly what it is
forms = filings["filings"]["recent"]["form"]
accessions = filings["filings"]["recent"]["accessionNumber"]
dates = filings["filings"]["recent"]["filingDate"]

for i, form in enumerate(forms):
    if form == "10-K":
        print("Raw accession:", accessions[i])
        print("Filing date:", dates[i])
        break
```

```python
# Step 2: clean it and build URL correctly
raw_accession = accessions[i]                    # e.g. 0000950170-24-129234
clean_accession = raw_accession.replace("-", "") # e.g. 000095017024129234
cik_no_zeros = str(int(cik))                     # e.g. 1318605 (no leading zeros)

index_url = f"https://www.sec.gov/Archives/edgar/data/{cik_no_zeros}/{clean_accession}/{clean_accession}-index.json"
print("URL:", index_url)

r = requests.get(index_url, headers=headers)
print("Status:", r.status_code)
```

```python
# Try .htm instead of .json
index_url = f"https://www.sec.gov/Archives/edgar/data/{cik_no_zeros}/{clean_accession}/{clean_accession}-index.htm"
r = requests.get(index_url, headers=headers)
print("Status:", r.status_code)

# Parse the htm index to find the main document
soup = BeautifulSoup(r.text, "html.parser")

# Find all links in the index page
for link in soup.find_all("a", href=True):
    href = link["href"]
    if href.endswith(".htm") and "10-k" in href.lower() or "tsla" in href.lower():
        print("Possible main doc:", href)
```

```python
# Most reliable approach — search EDGAR for Tesla's latest 10-K directly
search_url = "https://efts.sec.gov/LATEST/search-index?q=%22tesla%22&forms=10-K&dateRange=custom&startdt=2025-01-01&enddt=2026-12-31"
r = requests.get(search_url, headers=headers)
print("Status:", r.status_code)
print(r.text[:1000])
```
